# **Import Libraries**

```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Ensure the model ID is correct
model_id = "CompVis/stable-diffusion-v1-4"
```

```python
# Check if CUDA (GPU) is available and set device accordingly
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

# **Load Stable Diffusion**

```python
# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe.to(device)
```

# **User Input For Text Prompt**

```python
prompt = input("Enter a text prompt to generate an image: ")
```

```python
# Generate an image from the user-provided text prompt
with torch.autocast("cuda"):
    image = pipe(prompt).images[0]
```

```python
# Display the generated image
plt.imshow(image)
plt.axis('off')
plt.show()
```
